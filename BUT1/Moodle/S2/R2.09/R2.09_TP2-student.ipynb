## Calcul de la limite


On va maintenant essayer de calculer effectivement la limite de cette suite.

Pour cela on va utiliser des suites auxiliaires et un nouveau théorème



#### Suites adjacentes


 > Les suites $(u_n)_{n}$ et $(v_n)_{n}$ sont dites  adjacentes si
 > - $(u_n)_{n}$ est croissante et $(v_n)_{n}$ est décroissante,
 > - pour tout $n\geq 0$, on a $u_n\leq v_n$,
 > - $\lim_{n\to +\infty} (v_n -u_n) = 0$.


<div class="alert alert-block alert-info"><b>Théorème </b> 
Si les suites $(u_n)_{n}$ et $(v_n)_{n}$ sont adjacentes,
  elles convergent vers la même limite.

</div>


Il y a donc deux résultats dans ce théorème, la convergence de $(u_n)$ et
$(v_n)$ et en plus l'égalité des limites.
Les termes de la suites sont ordonnées ainsi :
$$u_0 \le u_1 \le u_2 \le \cdots \le u_n \le \cdots \cdots \le v_n\le \cdots \le v_2 \le v_1 \le v_0$$

***

#### Estimation de la limite

Soient $(b_n)$ et $(v_n)$ les deux suites définies pour $n \ge 1$ par
$$b_n = \sum_{k=1}^n \frac 1 {k^2} = 1+\frac{1}{2^2} +\frac{1}{3^2}+\cdots+\frac{1}{n^2} \quad \text{ et } \quad v_n = b_n + \frac2{n+1} .$$


**question 3**

Montrer que $(b_n)$ et $(v_n)$ sont deux suites adjacentes

Tracer sur la même figure les deux courbes.

En déduire une approximation de la limite de la suite $(b_n)_n$

```python
#TODO


def bale_up(n):
    return bale(n)+ 2/(n+1)


rg = range(1, 50)
s1 = [bale(n) for n in rg]
s2 = [bale_up(n) for n in rg]
plt.title('Suites adjacentes')
plt.plot(rg, s1, 'r+', markersize=2, label='$u_n$')
plt.plot(rg, s2, 'bx', markersize=2, label='$v_n$')

plt.legend(loc='best', numpoints=1, handlelength=3)
plt.show()
```
